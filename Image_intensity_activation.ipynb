```
from keras.datasets import mnist
import matplotlib.pyplot as plt
%matplotlib inline
```

Load the MNIST Data

```
(X_train,y_train), (X_test, y_test) = mnist.load_data()
```

Display a figure

```
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(1, 6, 1, xticks=[], yticks=[])
ax.imshow(X_train[7], cmap='gray')
ax.set_title(str(y_train[7]))
```

visualize one number with pixel intensities represented as activations

```
def visualize_input(img, ax):
    ax.imshow(img, cmap='gray')
    width, height = img.shape
    thresh = img.max()/2.5
    for x in range(width):
        for y in range(height):
            ax.annotate(str(round(img[x][y]/255.0,2)), xy=(y,x),
                        horizontalalignment='center',
                        verticalalignment='center',
                        color='white' if img[x][y]<thresh else 'black')
fig = plt.figure(figsize = (12,12)) 
ax = fig.add_subplot(111)
visualize_input(X_train[7], ax)
```
